```python
import pandas as pd

from siete import simulate_games_multithread, simulate_games_multiprocess

def get_simulation_df(n_games: int, n_players: int) -> pd.DataFrame:
    games = simulate_games_multiprocess(n_games, n_players)
    round_number = []
    pot = []
    winner = []
    for game in games:
        round_number.append(game.round_number)
        pot.append(game.pot)
        winner.append(game.winner.order)

    df = pd.DataFrame({'round_numbers': round_number, 'pot': pot, 'winner': winner})
    return df
```
